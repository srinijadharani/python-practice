**LINE GRAPH**

```
import matplotlib.pyplot as plt

x = [0,1,2,3,4,5,6,7,8]
y = [0,89,56,92,78,75,97,34,90]

plt.plot(x, y, color = "yellow", marker = 'o', markerfacecolor = "blue", markersize = 10)
plt.xlabel("Semester")
plt.ylabel("Grade")
plt.title("Result Analysis")
plt.show()
```

**BAR GRAPH**

```
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8]
y = [89,56,92,78,75,97,34,90]
plt.bar(x, y, tick_label = ["sem1", "sem2", "sem3", "sem4", "sem5", "sem6", "sem7", "sem8"], width = 0.45, 
        color = ["aquamarine", "antiquewhite"])
plt.xlabel("Semester")
plt.ylabel("Grade")
plt.title("Result Analysis")
plt.show()
```

**SCATTERED POINT GRAPH**

```
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8]
y = [89,56,92,78,75,97,34,90]
plt.scatter(x, y, label = "stars", color = "darkslateblue", marker = "*", s= 50)
plt.xlabel("Semester")
plt.ylabel("Grade")
plt.title("Result Analysis")
plt.show()
```

**PIE CHART**

```
import matplotlib.pyplot as plt
labels = ["Python", "Java", "C", "JavaScript"]
sizes = [75, 80, 100, 90]
colors = ["darkturquoise", "darkslateblue", "aqua", "darkcyan"]
explode = (0.1, 0, 0, 0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors)

plt.axis("equal")
plt.show()
```
